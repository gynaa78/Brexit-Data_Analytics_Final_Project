```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
referendum = pd.read_csv('referendum.csv')
referendum.head(10)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Avg Percent_Turnout per Region")

splot=sns.barplot(x ='Region', y='Percent_Turnout', data= referendum, estimator=np.mean) 
plt.xticks(rotation=45)

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=15,
                   xytext = (0, -12), 
                   textcoords = 'offset points')

plt.xlabel("Region")
plt.ylabel("Percent_Turnout")

plt.legend (['x=Region', 'y=Percent_Turnout'])

plt.savefig("add_annotation_to_bars_in_barplot_Seaborn_Python.png")





```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Rejected_Ballots per Region")

splot= sns.barplot(x='Region', y='Rejected_Ballots', data= referendum, estimator=np.mean) 
plt.xticks(rotation=45)

for p in splot.patches:
    splot.annotate(format(p.get_height(), '.1f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   size=15,
                   xytext = (0, -12), 
                   textcoords = 'offset points')


plt.xlabel("Region")
plt.ylabel("Rejected_Ballots")
plt.legend (['x=Region', 'y=Rejected_Ballots'])

plt.savefig("add_annotation_to_bars_in_barplot_Seaborn_Python.png")

plt.show()





    
```
